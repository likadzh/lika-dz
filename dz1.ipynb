```python
from bs4 import BeautifulSoup
import requests
import re
from pymystem3 import Mystem
import pymorphy2
m = Mystem()
from nltk.corpus import stopwords 
stopWords = set(stopwords.words('russian'))
from fake_useragent import UserAgent
ua = UserAgent(verify_ssl=False)
```

```python
session = requests.session()
slist = []
olist = []
urls = ['https://www.wildberries.ru/catalog/4545690/otzyvy', 
        'https://www.wildberries.ru/catalog/10141551/otzyvy',
        'https://www.wildberries.ru/catalog/5717207/otzyvy',
        'https://www.wildberries.ru/catalog/5761410/otzyvy',
        'https://www.wildberries.ru/catalog/5761401/otzyvy',
        'https://www.wildberries.ru/catalog/5283633/otzyvy']
for i in urls:
    req = session.get(i)
    req.encoding = 'utf-8'
    htmlString = req.text
    soup = BeautifulSoup(htmlString, 'html.parser')   
    s = soup.find_all("p", itemprop="reviewBody")
    for each in s: 
        slist.append(str(each.get_text()))
    o = soup.find_all("div", itemprop="ratingValue")
    for each in o: 
        olist.append(str(each.get_text()))
    #slist.append(s)
    #olist.append(o)
#print(slist)
#print(olist)
```

```python
qlist = []
req = session.get('https://www.wildberries.ru/catalog/8749436/otzyvy')
soup = BeautifulSoup(req.text, 'html.parser')
s = soup.find_all("p", itemprop="reviewBody")
for each in s: 
    qlist.append(str(each.get_text()))
```

```python
clear_qlist = []
for i in qlist:
    result = re.sub(r'[^–∞-—è—ë–ê-–Ø–Å\s]|\r|\n', ' ', i, flags=re.UNICODE)
    clear_qlist.append(result.lower())
#clear_qlist = clear_qlist[:10]
print (clear_qlist)
```

```python
l_qlemmas = []
for i in clear_qlist:
    qlemmas = m.lemmatize(i)
    qlemmas = ''.join(qlemmas)
    qlemmas = qlemmas.replace('\n', '')
    l_qlemmas.append(qlemmas)
print (l_qlemmas)
```
