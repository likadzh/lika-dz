```python
session = requests.session()
slist = []
olist = []
urls = ['https://www.wildberries.ru/catalog/4545690/otzyvy', 
        'https://www.wildberries.ru/catalog/10141551/otzyvy',
        'https://www.wildberries.ru/catalog/5717207/otzyvy',
        'https://www.wildberries.ru/catalog/5761410/otzyvy',
        'https://www.wildberries.ru/catalog/5761401/otzyvy',
        'https://www.wildberries.ru/catalog/5283633/otzyvy']
for i in urls:
    req = session.get(i)
    req.encoding = 'utf-8'
    htmlString = req.text
    soup = BeautifulSoup(htmlString, 'html.parser')   
    s = soup.find_all("p", itemprop="reviewBody")
    for each in s: 
        slist.append(str(each.get_text()))
    o = soup.find_all("div", itemprop="ratingValue")
    for each in o: 
        olist.append(str(each.get_text()))
    #slist.append(s)
    #olist.append(o)
#print(slist)
#print(olist)
```
